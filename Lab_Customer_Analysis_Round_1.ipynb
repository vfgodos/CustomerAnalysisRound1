```python
file3 = file3.rename(columns={'State':'ST', 'Gender':'GENDER'})
file3.head()
```

```python
file2 = file2[['Customer', 'ST', 'GENDER', 'Education', 
               'Customer Lifetime Value', 'Income', 
               'Monthly Premium Auto', 'Number of Open Complaints', 
               'Policy Type', 'Vehicle Class', 'Total Claim Amount']]
               
               
               
```

```python
file3 = file3[['Customer', 'ST', 'GENDER', 'Education', 
               'Customer Lifetime Value', 'Income', 
               'Monthly Premium Auto', 'Number of Open Complaints', 
               'Policy Type', 'Vehicle Class', 'Total Claim Amount']]
```

```python
data = pd.concat([file1, file2, file3], axis=0)
data.shape
data
```

```python
data.dtypes
```

```python
#Numerical columns: Income, Monthly Premium Auto, Total Claim Amount, Number of Open Complaints, Customer, Customer Lifetime Value
#Categorical columns: ST, Gender, Education, policy type, vehicle class
#Meaning:
#Customer: customer identificacion
#ST: customer state
#Gender: customer gender
#Education: customer grade of education
#Customer Lifetime Value: the value that the client has for the brand
#Income: amount of income
#Monthly Premium Auto: premium payment amount
#Number of Open Complaints: Complaints that have not been solved
#Policy Type: type of policy that the client has 
#Vehicle Class: type of vehicle
#Total Claim Amount: quantity of the claim required by the client
```

```python
data_new = data.drop(['Education','Number of Open Complaints'], axis=1)
data_new.head()
```
